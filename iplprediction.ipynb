```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()), list(ipl['player_of_match'].value_counts()[0:5].keys()))
plt.show()        
```

```python
ipl['result'].value_counts()
```

```python
ipl['toss_winner'].value_counts()
```

```python
batting_first = ipl[ipl['target_runs'] !=0]
```

```python
batting_first.head()
```

```python
plt.figure(figsize = (5,7))
plt.hist(batting_first['target_runs'])
plt.title("Distribution of Runs")
plt.xlabel("runs")
```

```python
batting_first['winner'].value_counts()
```

```python
plt.figure(figsize = (8,12))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()), list(batting_first['winner'].value_counts()[0:3]),color = ["blue", "black", "pink"])
plt.show()
```

```python
plt.figure(figsize = (7,7))
plt.pie(list(batting_first['winner'].value_counts()), labels = list(batting_first['winner'].value_counts().keys()), autopct = '%0.1f%%')
```

```python
batting_second = ipl[ipl['target_overs']!=0]
```

```python
batting_second.head()
```

```python
plt.figure(figsize = (7,7))
plt.hist(batting_second['target_overs'], bins = 30)
plt.show()
```

```python
batting_second['winner'].value_counts()
```

```python
plt.figure(figsize = (8,12))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()), list(batting_second['winner'].value_counts()[0:3]),color = ["yellow", "orange", "purple"])
plt.show()
```

```python
plt.figure(figsize = (8,12))
plt.pie(list(batting_second['winner'].value_counts()), labels = list(batting_second['winner'].value_counts().keys()), autopct = '%0.1f%%')
```
